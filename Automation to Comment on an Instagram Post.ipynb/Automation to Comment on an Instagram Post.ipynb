```python
import pyautogui  # do the automation of the mouse and keyboard
import pyperclip  # used to copy and paste
import time  # controls the time of our program

# ALERT: This program will always comment the latest post of the profile informed by the user.

profile = input("Type the instagram profile's name you want to comment at: ")
comment = input('Type the comment: ')
number = int(input('Type how many times you want to comment the text informed above: '))

counter = 0

# it opens a new tab on Google Chrome
pyautogui.hotkey('ctrl', 't')

# it waits two seconds  
time.sleep(2)
    
# it types "instagram.com" in the search box of Google
pyperclip.copy('instagram.com')
pyautogui.hotkey('ctrl', 'v')

# it waits two more seconds
time.sleep(2)

# it presses "enter" to go to instagram
pyautogui.press('enter')

# it waits eight seconds
time.sleep(8)

# it clicks on the instagram's search box
pyautogui.click(959, 176)
    
# it waits two seconds
time.sleep(2)

# it types the name of the profile informed above
pyperclip.copy(profile)
pyautogui.hotkey('ctrl', 'v')

# it waits six seconds
time.sleep(6)
 
# it clicks on the first profile option
pyautogui.click(870, 297)
    
# it waits eight seconds
time.sleep(8)
    
# it clicks on the latest publication
pyautogui.click(1909, 904)
time.sleep(3)
pyautogui.click(487, 413)
    
# it waits seven seconds
time.sleep(7)

# the loop will start and only ends when all the comments are posted
while counter != number:

    time.sleep(2)  # it waits two seconds

    pyautogui.click(1303, 993)  # it clicks on the commentary box
    
    time.sleep(3)  # it waits three seconds
    
    # it types the comment informed above
    pyperclip.copy(comment)
    pyautogui.hotkey('ctrl', 'v')
    
    time.sleep(5)  # it waits five seconds
    
    pyautogui.press('enter')  # it presses "enter" in order to publish the comment
    
    time.sleep(3)  # it waits three seconds
    
    pyautogui.alert(' Another comment successfully posted!')  # it alerts the user that the comment is posted
    
    time.sleep(4)  # it waits four seconds

    counter = counter + 1  # increments the counter
    
pyautogui.alert(' The process is completed. To finish it, click "Ok".')  # it alerts the user that the process is finished
```
